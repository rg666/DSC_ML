```python
# Loading the necessary libraries
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset_spine.csv")


def line_break():
    '''Provide a custom line break string for good visibility'''
    print('******************************************************************************************************')
```

# Basic data exploration

```python
display(df.head(10))
line_break()
print(df.shape)
line_break()
display(df.info())
line_break()
display(df.describe())
line_break()
print(df.isnull)
```

Drop the NaNs column Unnamed: 13 and rename to columns and set Status to categorical column.

```python
del df['Unnamed: 13']
df.columns = ['pelvic_incidence', 'pelvic tilt', 'lumbar_lordosis_angle', 'sacral_slope', 'pelvic_radius',
              'degree_spondylolisthesis', 'pelvic_slope', 'Direct_tilt', 'thoracic_slope', 'cervical_tilt', 'sacrum_angle', 'scoliosis_slope', 'Status']

display(df.head(10))
df['Status'] = pd.Categorical(df.Status)
```

```python
sns.countplot(x='Status', data=df)
plt.show()
```

```python
'''Plot the pairplots'''
sns.pairplot(df)
```

Correlation map

```python
# sns.heatmap(df.corr(), annot=True, fmt=".2f")
# plt.show()

corr_back = df.corr()

# Generate the correlation plot
sns.heatmap(corr_back, center=0, square=True, linewidths=.5)
plt.show()
line_break()
'''Correlation Values'''
display(corr_back)
line_break()
```

```python
display(df.groupby('Status').mean())
line_break()
display(df.groupby('Status').median())
```

```python
# Generating 3*4 matrix of box plots
fig, axes = plt.subplots(3, 4, figsize=(15, 15))
axes = axes.flatten()

for i in range(0, len(df.columns)-1):
    sns.boxplot(
        x="Status", y=df.iloc[:, i], data=df, orient='v', ax=axes[i])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
ax = sns.scatterplot(x="pelvic_radius", y="degree_spondylolisthesis",
                     hue="Status", data=df, palette=['red', 'green'])
plt.show()
```
